```python
df = pd.read_csv(dataset_path, index_col=0, dtype={'patient': 'str'})
```

```python
for center in range(5):
    samples_in_center = len(df[df['center'] == center])
    tumor_samples_in_center = len(df[(df['center'] == center) & (df['tumor'] == 1)])
    non_tumor_samples_in_center = len(df[(df['center'] == center) & (df['tumor'] == 0)])
    print(f'Center {center}: Total samples = {samples_in_center}, Tumor samples = {tumor_samples_in_center}, Non-tumor samples = {non_tumor_samples_in_center}')
```
