```python
import pandas as pd
```

```python
df =  pd.read_csv('metadata.csv', index_col=0, dtype={'patient': 'str'})
```

```python
for center in df['center'].unique():
    subset = df[df["center"] == center]
    print(f"For hospital {center}, the patients are {subset['patient'].unique()}. ")
```

```python
for patient in df['patient'].unique():
    # how many rows corresponding to this patient
    subset = df[df["patient"] == str(patient)]
    print(f"For patient {patient}, there are {len(subset)} entries.")
```

```python
patients_to_keep = [
    # Hospital 0
    '009',  # train
    '010',  # val
    
    # Hospital 1
    '036',  # train
    '024',  # val
    
    # Hospital 2
    '042',  # train
    '044',  # val
    
    # Hospital 3
    '067',  # train
    '062',  # val
    
    # Hospital 4
    '087',  # train
    '092'   # val
]
```

```python
filtered_df = df[df['patient'].isin(patients_to_keep)]
```
